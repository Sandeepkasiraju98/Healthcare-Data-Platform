```python
# Access Doctor collection
doctor_collection = db['Doctor']

# Find the last used DoctorID, if any
last_doctor = doctor_collection.find_one(sort=[("DoctorID", -1)])

# Determine the starting value for DoctorID
start_doctor_id = 1 if last_doctor is None else last_doctor['DoctorID'] + 1

# Initialize doctor_id with start_doctor_id
doctor_id = start_doctor_id

# Data for Doctor
doctor_data = [
    {
        "DoctorID": doctor_id,
        "EmployeeID": 1,
        "Specialization": "Cardiology"
    },
    
    {
        "DoctorID": doctor_id,
        "EmployeeID": 2,
        "Specialization": "Neurology"
    },
    
    {
        "DoctorID": doctor_id,
        "EmployeeID": 4,
        "Specialization": "Oncology"
    },
    {
        "DoctorID": doctor_id,
        "EmployeeID": 1,
        "Specialization": "Orthopedics"
    },
    {
        "DoctorID": doctor_id,
        "EmployeeID": 2,
        "Specialization": "General Surgery"
    },
    
    {
        "DoctorID": doctor_id,
        "EmployeeID": 4,
        "Specialization": "Psychiatry"
    }
]

# Insert the data into Doctor collection
for data in doctor_data:
    employee_id = data["EmployeeID"]
    specialization = data["Specialization"]
    
    # Check if there is an existing record with the same EmployeeID and Specialization
    existing_doctor = doctor_collection.find_one({"EmployeeID": employee_id, "Specialization": specialization})
    
    if existing_doctor:
        print(f"Found existing DoctorID {existing_doctor['DoctorID']} for EmployeeID {employee_id} and Specialization {specialization}.")
        continue
    
    # Check if EmployeeID exists in Nurse or GeneralStaff collection
    nurse_exists = db['Nurse'].find_one({"EmployeeID": employee_id})
    generalstaff_exists = db['GeneralStaff'].find_one({"EmployeeID": employee_id})
    
    if nurse_exists or generalstaff_exists:
        print(f"Error: EmployeeID {employee_id} already exists in Nurse or GeneralStaff table. Skipping insertion.")
        continue
    
    # Check if EmployeeID exists in Employee collection
    employee_exists = db['Employee'].find_one({"EmployeeID": employee_id})
    if not employee_exists:
        print(f"Error: EmployeeID {employee_id} not found in Employee collection. Skipping insertion.")
        continue
    
    # Insert the data into Doctor collection with existing or auto-incremented DoctorID
    doctor_id = existing_doctor['DoctorID'] if existing_doctor else start_doctor_id
    data["DoctorID"] = doctor_id
    result = doctor_collection.insert_one(data)
    print("Inserted ID:", result.inserted_id)
    
    # Increment DoctorID for the next entry only if it's a new insertion
    if not existing_doctor:
        start_doctor_id += 1

# Output message after all data insertion
print("All entries inserted successfully.")
```
